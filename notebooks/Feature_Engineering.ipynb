```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
import math 
from pathlib import Path
from scipy import stats
```

```python
raw_data = pd.read_csv(f"{Path.cwd().parent}/data/raw/delivery_data.csv")
```

#### Converting datetime to a flag suggesting if its peak hour or not

```python
def peak_hour(hour):
    if (hour >=12 and hour <=13) or (hour >=18 and hour<=20):
        return "High"
    else:
        return "Normal"
```

```python
raw_data['time_of_day'] = raw_data['hour'].apply(peak_hour)
```

```python
raw_data['weather_traffic'] = raw_data['weather']+"_"+raw_data['traffic_level']
```

```python
raw_data['asp'] = round((raw_data['order_value']/raw_data['num_items']).astype(float),2)
```

```python
raw_data = raw_data.drop(['order_value'],axis = 1)
```
